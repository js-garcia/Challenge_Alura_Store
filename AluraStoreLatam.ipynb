### Importación de datos



```python
import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

tienda.head()
```

#1. Análisis de facturación



```python


facturacion_tienda1 = tienda['Precio'].sum()
facturacion_tienda2 = tienda2['Precio'].sum()
facturacion_tienda3 = tienda3['Precio'].sum()
facturacion_tienda4 = tienda4['Precio'].sum()

facturacion_tienda1, facturacion_tienda2, facturacion_tienda3, facturacion_tienda4
```

```python
facturacion = pd.DataFrame({
    'Tienda': ['Tienda 1', 'Tienda 2', 'Tienda 3', 'Tienda 4'],
    'Facturación Total': [facturacion1, facturacion2, facturacion3, facturacion4]
})

facturacion.sort_values('Facturación Total', ascending=False)
```

## **Visualizacion**

#### Grafico 1 - Ingresos por tienda

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(8,5))

plt.bar(
    ['Tienda 1', 'Tienda 2', 'Tienda 3', 'Tienda 4'],
    [facturacion_tienda1, facturacion_tienda2, facturacion_tienda3, facturacion_tienda4],
    color='skyblue'
)

plt.title('Facturación total por tienda (R$)')
plt.ylabel('Reales (R$)')
plt.xlabel('Tiendas')

plt.show()
```

### Grafico 2 - Ventas por categoria

```python
plt.figure(figsize=(8,6))

plt.pie(
    ventas_categoria_t1,
    labels=ventas_categoria_t1.index,
    autopct='%1.1f%%',
    startangle=90
)

plt.title('Distribución de ventas por categoría - Tienda 1')
plt.show()
```

### Grafico 3 - Calificacion promedio por tienda

```python
plt.figure(figsize=(8,5))

plt.bar(
    ['Tienda 1', 'Tienda 2', 'Tienda 3', 'Tienda 4'],
    [calificacion_promedio_t1, calificacion_promedio_t2, calificacion_promedio_t3, calificacion_promedio_t4],
    color='lightgreen'
)

plt.title('Calificación promedio por tienda')
plt.ylabel('Puntaje (0–5)')
plt.xlabel('Tiendas')
plt.ylim(0, 5)   # porque las calificaciones van de 0 a 5

plt.show()
```

## **Informe Final - Analisi de Tiendas de Alura Store**

#### El objetivo del análisis fue determinar qué tienda de Alura Store debería vender el Sr. Juan, evaluando ingresos, ventas por categoría, calificaciones, productos vendidos y costos de envío. A partir de estos datos, se comparó el rendimiento de las cuatro tiendas.

### Resumen de analisis


1. Ingresos totales

Los ingresos muestran diferencias claras entre tiendas.

* La tienda con mayor recaudación es Tienda 1.

* La tienda con menor recaudación es Tienda 4, con 1.038.376.000.

* Esto indica un menor desempeño financiero de Tienda 4.

2. Ventas por categoría

* Las tiendas con mejor rendimiento muestran categorías con alta rotación.
* Tienda 4 presenta menor movimiento en varias categorías, indicando baja demanda.

3. Calificación promedio

Las calificaciones son muy similares entre tiendas, sin diferencias significativas.
Este indicador no afecta de manera importante la decisión final.

4. Productos más y menos vendidos

Tienda 4 muestra más productos con baja cantidad de ventas, lo que evidencia menor aceptación del público y problemas de rotación.

5. Costo de envío promedio

Tienda 4 presenta un costo de envío más alto en comparación con las otras tiendas, lo cual impacta negativamente porque el cliente paga el envío.

## **Recomendacion**


El Sr. Juan debería vender Tienda 4, ya que es la que menos aporta al rendimiento global de la cadena y presenta más debilidades en los indicadores evaluados.
